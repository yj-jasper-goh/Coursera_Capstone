### Scrap wikipedia page for the table of postal code of Toronto

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
table1 = str(soup.table)
display_html(table1,raw=True)
```

### The html "table1" from the Wikipedia page to be converted to a Pandas Data Frame

```python
dfs = pd.read_html(table1)
df=dfs[0]
df.head(20)
```

```python
df.shape
```

### Remove 'Not assigned' rows and reset index

```python
df.drop(df[df['Borough'] == 'Not assigned'].index, axis = 0, inplace = True)
df.drop(df[df['Neighbourhood'] == 'Not assigned'].index, axis = 0, inplace = True)
```

```python
df.rename(columns={'Postal Code':'PostalCode'},inplace=True)
df.reset_index(drop=True).head(20)
```
