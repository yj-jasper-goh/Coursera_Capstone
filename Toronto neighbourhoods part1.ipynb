```python
df = pd.read_excel('https://torontocapstone.s3.jp-tok.cloud-object-storage.appdomain.cloud/Toronto%20neighbourhoods.xlsx',
                       sheet_name='Sheet1',
                       skiprows=range(0),
                       skipfooter=0
                      )

print('Data downloaded and read into a dataframe!')
```

Step 2 \
Remove 'Not assigned' rows and reset index

```python
df.drop(df[df['Borough'] == 'Not assigned'].index, axis = 0, inplace = True)
df.drop(df[df['Neighbourhood'] == 'Not assigned'].index, axis = 0, inplace = True)
```

```python
df.rename(columns={'Postal Code':'PostalCode'},inplace=True)
df.reset_index(drop=True).head(20)
```
